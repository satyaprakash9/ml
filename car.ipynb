```python
car_data = car_data.drop('car_ID',axis=1)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(car_data.corr(),cmap='RdYlGn',annot=True)
```

```python
car_corr = car_data.corr()['price']
top_features = car_corr[abs(car_corr)>0.5].index.tolist()
```

```python
top_features
```

```python
for col in car_data.select_dtypes([float,int]).columns:
  if col not in top_features:
    car_data = car_data.drop(col,axis=1)
car_data.head()
```

```python
car_data.select_dtypes([float,int]).columns
```

```python
for col in car_data.select_dtypes([float,int]).columns:  
  desc = car_data[col].describe()
  iqr = desc['75%']-desc['25%']
  max = iqr*1.5 + desc['75%']
  car_data = car_data.drop(index=car_data[col][car_data[col]>max].index)
```

```python
from sklearn.preprocessing import StandardScaler
s = StandardScaler()
for col in car_data.select_dtypes([float,int]).columns:
  if col != 'price':
    car_data[col] = s.fit_transform(car_data[[col]])
```

```python
car_data = pd.get_dummies(car_data,columns=car_data.select_dtypes(object).columns,drop_first=True)
```

```python
from scipy.stats import boxcox,normaltest
price_bc = boxcox(car_data['price'])
```

```python
price_log = np.log1p(car_data['price'])
```

```python
normaltest(price_bc[0])
```

```python
normaltest(price_log)
```

```python
car_data.price = price_log
```

```python
X = car_data.drop('price',axis=1)
y = car_data['price']
```

```python
car_data.price.skew()
```

```python
sns.histplot(data=car_data,x='price')
```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV, ElasticNetCV, Lasso
from sklearn.model_selection import cross_val_predict, KFold, train_test_split, GridSearchCV
from sklearn.metrics import r2_score
pf = PolynomialFeatures(degree=2)
s = StandardScaler()
lr = LinearRegression()
kf = KFold(shuffle=True,random_state=8457,n_splits=3)
estimator = Pipeline([('polynomial_features',pf),('linear_regression',lr)])
predictions = cross_val_predict(estimator,X,y,cv=kf)
```

```python
pred = cross_val_predict(lr,X,y,cv=kf)
```

```python
plt.plot(np.exp(y)-1,np.exp(y)-1)
```

```python
sns.scatterplot(np.exp(y)-1,np.exp(pred)-1)
plt.plot(np.exp(y)-1,np.exp(y)-1)
plt.ylim((np.exp(y)-1).min(),(np.exp(y)-1).max())
plt.xlim((np.exp(y)-1).min(),(np.exp(y)-1).max())
plt.xlabel('expected values')
plt.ylabel('predicted values')
```

```python
r2_score(y,pred)
```

```python
r2_score(y,predictions)
```

```python
plt.scatter(np.exp(y)-1,np.exp(predictions)-1)
plt.plot(np.exp(y)-1,np.exp(y)-1)
plt.xlim((np.exp(y)-1).min(),(np.exp(y)-1).max())
plt.ylim((np.exp(y)-1).min(),(np.exp(y)-1).max())
plt.xlabel('expected values')
plt.ylabel('predicted values')
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.3,random_state=8457)
alphas = [1e-5, 5e-5, 0.0001, 0.0005]
lascv = LassoCV(alphas=alphas,max_iter=100000,cv=kf).fit(X_train,y_train)
r2_score(y_test,lascv.predict(X_test))
```

```python
(lascv.coef_==0).sum()
```

```python
lascv.alpha_
```

```python
sns.scatterplot(np.exp(y)-1,np.exp(lascv.predict(X))-1,label='predict',color='r')
sns.lineplot(np.exp(y)-1,np.exp(y)-1,label='real')
plt.ylim((np.exp(y)-1).min(),(np.exp(y)-1).max())
plt.xlim((np.exp(y)-1).min(),(np.exp(y)-1).max())
plt.xlabel('expected values')
plt.ylabel('predicted values')
```

```python
ridcv = RidgeCV(alphas=[0.005, 0.05, 0.1, 0.3, 1, 3, 5, 10, 15, 30, 80],cv=kf).fit(X_train,y_train)
r2_score(y_test,ridcv.predict(X_test))
```

```python
ridcv.alpha_
```

```python
plt.scatter(np.exp(y)-1,np.exp(ridcv.predict(X)))
plt.xlabel('expected values')
plt.ylabel('predicted values')
```

```python
estimator1 = Pipeline([('poly_features',PolynomialFeatures()),('lasso',Lasso())])
params = {'poly_features__degree':[1,2,3],'lasso__alpha':[1e-5, 5e-5, 0.0001, 0.0005]}
grid = GridSearchCV(estimator1,params,cv=kf).fit(X_train,y_train)
r2_score(y_test,grid.predict(X_test))
```

```python
elastic_cv = ElasticNetCV(alphas=alphas,l1_ratio=np.linspace(0,1,10),max_iter=100000).fit(X_train,y_train)
r2_score(y_test,elastic_cv.predict(X_test))
```
